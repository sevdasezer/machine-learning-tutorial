```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://github.com/yigitsener/machine-learning/blob/master/data/Fish.csv" ## veriye github üzerinden ulaşmak
df = pd.read_csv(url + "?raw=true", error_bad_lines=False)
```

```python
df.shape # satır ve değişken sayısı
```

```python
df.head()
```

```python
df.Species.value_counts()
```

```python
df.Length1.describe()
```

```python
sns.violinplot(y = df.Length1, linewidth=5)# seaborn ile violin plot oluşturma
plt.show() # sadece Lenght1 değişkeni için grafik oluşturulacak
```

df.Length1.describe() çıktısında yer alan sonuçları incelediğimizde medyan ve ortalama arasında birbirine yakın ve çeyrekler ile medyan arasında da hemen hemen yakın bir dağılım olduğunu görmekteyiz. Ancak 3. çeyrek sonrasında maksimum değere yaklaşan sayıların minimuma yaklaşanlara göre fazlalaştığı görülmektedir. Aynı şekilde violin plottta da görüldüğü üzere yukarı doğru yoğunluğun azaldığı görülmektedir. Yani temel istatistikler yoğunluk bazında violin plot üzerinden görebilmekteyiz.

```python
# sadece Perch ve Bream türlerini seçelim
df_two = df[(df["Species"] == "Perch") | (df["Species"] == "Bream")]

# Violin plot üzerinden
# uzunlukları (Length1) karşılaştıralım
sns.violinplot(x="Species", y="Length1", data=df_two, linewidth=5)
plt.show()
```

```python
sns.violinplot(x="Length1", y="Species", data=df_two, linewidth=5)
plt.show()
```

```python
sns.violinplot(x="Species", y="Length1", data=df, linewidth=3)
plt.show()
```

```python
sns.violinplot(x="Species", y="Length1", data=df, palette="Set2")
plt.show()
```
